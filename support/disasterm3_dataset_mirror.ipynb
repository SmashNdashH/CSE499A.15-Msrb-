```python
import os
import json
from huggingface_hub import hf_hub_download, snapshot_download
from kaggle_secrets import UserSecretsClient

# 1. Authenticate Kaggle API using your attached Secrets
user_secrets = UserSecretsClient()
os.environ["KAGGLE_USERNAME"] = user_secrets.get_secret("KAGGLE_USERNAME")
os.environ["KAGGLE_KEY"] = user_secrets.get_secret("KAGGLE_KEY")

# 2. Setup /tmp directory
LOCAL_DIR = "/tmp/disasterm3"
os.makedirs(LOCAL_DIR, exist_ok=True)
REPO_ID = "Kingdrone-Junjue/DisasterM3"

# 3. Download the large zips and manifests from Hugging Face
print("Downloading Instruct Set from HF...")
instruct_files = [
    "DisasterM3_Instruct/train_images.zip",
    "DisasterM3_Instruct/box_train_images.zip",
    "DisasterM3_Instruct/masks.zip",
    "DisasterM3_Instruct/train_release.json",
]
for f in instruct_files:
    hf_hub_download(repo_id=REPO_ID, filename=f, repo_type="dataset", local_dir=LOCAL_DIR)


# (We are intentionally skipping the Bench set images as they cause HF downloads to hang 
# and are not needed for fine-tuning. We will handle them during the evaluation phase.)
# print("Downloading Bench Set from HF...")
# hf_hub_download(repo_id=REPO_ID, filename="DisasterM3_Bench/benchmark_release.json", 
#                 repo_type="dataset", local_dir=LOCAL_DIR)

# # Also download the bench images/masks (requires snapshot_download for folders)
# snapshot_download(repo_id=REPO_ID, allow_patterns="DisasterM3_Bench/*", 
#                   repo_type="dataset", local_dir=LOCAL_DIR)

# 4. Initialize Kaggle Dataset metadata
print("Initializing Kaggle Dataset metadata...")
dataset_slug = "disasterm3-mirror" # Change this if the name is taken globally
metadata = {
  "title": "DisasterM3 Mirror",
  "id": f"{os.environ['KAGGLE_USERNAME']}/{dataset_slug}",
  "licenses": [{"name": "CC-BY-NC-SA-4.0"}]
}
with open(f"{LOCAL_DIR}/dataset-metadata.json", "w") as f:
    json.dump(metadata, f)

# 5. Push to Kaggle Datasets
print("Pushing to Kaggle Datasets (this will take a while)...")
!kaggle datasets create -p {LOCAL_DIR} --dir-mode zip
print("Done! Your permanent dataset is ready.")
```
